## VSCode로 실행할 때

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common import exceptions  
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.get('https://www.airportal.go.kr/life/airinfo/RbHanFrmMain.jsp#') 
driver.implicitly_wait(5)


month = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']

for m in month:
    if m == '02':
        end_day = 28
    elif m in ['04', '06', '09', '11']:
        end_day = 30
    else:
        end_day = 31

    monthly = pd.DataFrame()

    for d in range(1, end_day + 1):
        if d in range(1, 10):
            d = '0' + str(d)
        else:
            d = str(d)

        driver.get('https://www.airportal.go.kr/life/airinfo/RbHanFrmMain.jsp#')    
        driver.implicitly_wait(5)
        # 구분, 날짜 지정
        driver.switch_to.frame('main')
        # 구분을 '출발'로 변경
        departure_btn = driver.find_element_by_css_selector('input[type=radio]:nth-child(3)')
        departure_btn.click()
        # 날짜 선택하기
        calender = driver.find_element_by_css_selector('#current_date')
        calender.clear()
        date = f'2017{m}{d}'
        calender.send_keys(date)
        # 검색버튼
        search_btn = driver.find_element_by_css_selector('table > tbody > tr:nth-child(2) > td:nth-child(4) > a')
        search_btn.click()
        driver.implicitly_wait(5)

        day_flights = []
        driver.switch_to.frame('sframe')

        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        table = soup.select('body > form > table > tbody > tr > td > table > tbody > tr')

        for tr in table[:len(table)-1:2]:
            td = tr.select('td')
            reason = ''
            tip = td[14].a
            if tip:
                reason = tip['onmouseover'][11:-2]
            day_flights.append([date, td[0].text.strip(), td[2].text, td[4].text, td[6].text, td[10].text, td[12].text, td[14].text, reason])

        df = pd.DataFrame(day_flights, columns=['date', 'airline', 'flightNo', 'togo', 'originTime', 'departTime', 'type', 'state', 'reason'])
        monthly = pd.concat([monthly, df])
        print(len(monthly))
    # 월별 csv 파일 만들기
    monthly.to_csv(f'2017{m}.csv', index=False)
# 드라이버 종료
driver.close()
```
